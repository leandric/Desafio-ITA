```python
df = []
i = 0
for ativo in ativos:
    i=i+1

    print(i, ativo)
    base = yf.download(ativo, start = '2015-01-01', interval = '1d')['Close']
    base = pd.DataFrame(base)
    base['ativo'] = ativo
    #base.reset_index(inplace=True)
    df.append(base)
```

```python
df1 = pd.concat(df)
df1['data'] = pd.to_datetime(df1.index)
df1['dia'] = pd.to_datetime(df1['data'], format="%Y-%m-%d").dt.date
df1 = df1[['dia','Close','ativo']]
```

```python
ativos = df1['ativo'].unique()

df2 = df1[df1['ativo']==ativo][['Close']]
for ativo in ativos:
    df2[ativo] = df1[df1['ativo']==ativo][['Close']]
```

```python
del df2['Close']
```

```python
df2.reset_index(inplace=True)
df1.reset_index(inplace=True)
#df2['Date'] = pd.to_datetime(df2['Date'], format="%Y-%m-%d").dt.date 
df2.rename(columns={'Date':'Dia'}, inplace=True)
df2.to_excel('bases/base normal.xlsx', index=None)
pickle.dump(df2)
df2
```
